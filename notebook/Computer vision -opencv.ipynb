hierarchy: Optional output vector describing the hierarchical structure of the contours. It is used when you want to analyze relationships between contours.

### Histogram

Gives high level intuition of the pixel distribution in the image.

In OpenCV, a histogram, in simple terms, is a graphical representation of the distribution of pixel intensities in an image. It shows how many pixels have a specific brightness or color value. Histograms are often used for image analysis and processing to understand the image's overall characteristics and make adjustments or enhancements.

Here's a simplified explanation:

Imagine you have a black-and-white image, and you want to know how many pixels in the image are very dark, how many are moderately dark, and how many are very bright. A histogram will give you a graph where the x-axis represents the range of pixel values (from dark to bright), and the y-axis represents the number of pixels with those values. So, for a dark image, you might see a peak on the left side of the histogram, and for a bright image, you'd see a peak on the right side.

For color images, there are separate histograms for each color channel (red, green, and blue in the case of RGB images) to show the distribution of colors in the image.

Histograms can be useful in tasks like image contrast adjustment, thresholding, and various image processing techniques where understanding and manipulating pixel intensity distributions is important.

### Simple Thresholding

Simple thresholding is a fundamental image processing technique used to segment an image into two regions based on pixel intensity values: foreground and background. It's a straightforward method that assigns a binary value (usually 0 for black and 255 for white) to each pixel in the image based on a predefined threshold value.

Choose a Threshold Value: You begin by selecting a threshold value, often denoted as T. This value determines the separation point between the foreground and background in the image.

Segmentation: For each pixel in the image, you compare its intensity value to the threshold value:

If the pixel intensity is greater than or equal to the threshold value, the pixel is assigned to the foreground (usually white).
If the pixel intensity is less than the threshold value, the pixel is assigned to the background (usually black).
Binary Image: After comparing and assigning each pixel, you obtain a binary image, where pixels are either white or black.

Simple thresholding is commonly used for tasks such as binarization, which is essential for image segmentation, object detection, and character recognition. It's a basic technique and is often the first step in more advanced image processing tasks. However, its effectiveness can be limited in cases where the lighting conditions are uneven or when the objects of interest have varying intensities. In such cases, more advanced thresholding methods or adaptive thresholding techniques may be required.

retval, threshold = cv2.threshold(src, thresh, maxval, type[, dst])

src: This is the source image on which you want to apply thresholding. It should be a grayscale image (single-channel).

thresh: The threshold value. This is the pixel intensity value used to classify pixels as either foreground or background based on the comparison with this value.

maxval: The maximum value used to assign to pixels that are greater than or equal to the threshold value. Typically, this is set to 255 for white in binary images.

type: This parameter specifies the type of thresholding to apply, and it should be one of the following constants:

cv2.THRESH_BINARY: Pixels with intensity >= thresh are set to maxval; others are set to 0.

v2.THRESH_BINARY_INV: Inverse of cv2.THRESH_BINARY; pixels with intensity >= thresh are set to 0; others are set to maxval.

cv2.THRESH_TRUNC: Pixels with intensity >= thresh are set to thresh; others remain unchanged.

cv2.THRESH_TOZERO: Pixels with intensity >= thresh remain unchanged; others are set to 0.

cv2.THRESH_TOZERO_INV: Inverse of cv2.THRESH_TOZERO; pixels with intensity >= thresh are set to 0; others remain unchanged.

### Adaptive Thresholding

Adaptive thresholding is an image processing technique used to perform local thresholding on an image. Unlike global thresholding, where a single threshold value is applied to the entire image, adaptive thresholding calculates a unique threshold for each pixel based on its local neighborhood. This method is particularly useful when the lighting conditions vary across the image or when there's a need to segment objects with different levels of illumination. Here's how adaptive thresholding works in simple terms:

Choose a Neighborhood: You specify a local neighborhood around each pixel in the image. This neighborhood can be a small rectangular or square region.

Calculate a Threshold for Each Pixel: For each pixel in the image, a threshold is calculated based on the pixel values within its local neighborhood. Common methods for calculating the local threshold include mean (average) and Gaussian-weighted mean.

Compare Pixel Intensity to Local Threshold: Each pixel's intensity value is compared to its corresponding local threshold. The pixel is assigned to the foreground if its intensity is greater than or equal to the local threshold, and to the background otherwise.

Create a Binary Image: After comparing each pixel, you obtain a binary image where pixels are either classified as foreground or background.

cv2.adaptiveThreshold(src, maxValue, adaptiveMethod, thresholdType, blockSize, C[, dst])

rc: This is the source image (in grayscale) on which you want to apply adaptive thresholding.

maxValue: The maximum pixel value assigned to pixels that meet the threshold criteria. Typically, this is set to 255 for white in binary images.

adaptiveMethod: This parameter specifies the method used to calculate the local threshold for each pixel. It should be one of the following constants:

cv2.ADAPTIVE_THRESH_MEAN_C: Calculates the threshold as the mean of the local neighborhood.
cv2.ADAPTIVE_THRESH_GAUSSIAN_C: Calculates the threshold as a weighted sum of the local neighborhood values, where weights are based on a Gaussian window.
thresholdType: This parameter specifies the type of thresholding to apply locally, and it should be one of the following constants:

cv2.THRESH_BINARY: Pixels with intensity >= local threshold are set to maxValue; others are set to 0.

cv2.THRESH_BINARY_INV: Inverse of cv2.THRESH_BINARY; pixels with intensity >= local threshold are set to 0; others are set to maxValue.
blockSize: The size of the local neighborhood around each pixel. It should be an odd integer greater than 1.

C: A constant value subtracted from the mean (for cv2.ADAPTIVE_THRESH_MEAN_C) or the weighted sum (for cv2.ADAPTIVE_THRESH_GAUSSIAN_C) to adjust the threshold.

dst (optional): This is an optional output array (image) where the adaptive thresholded image will be stored. If not provided, the function returns the thresholded image.

### Edge detection

Laplacian:

!! Note this
Laplacian edge detection is a popular image processing technique used to detect edges in digital images. It is based on the second derivative of the image and highlights regions of rapid intensity change. The Laplacian operator is particularly useful for identifying edges because it responds to the rate of change in intensity at a given pixel location.

Here's how Laplacian edge detection works:

1. **Grayscale Conversion**: To perform edge detection, the first step is often to convert the color image to grayscale. This simplifies the problem to working with a single channel representing intensity.

Sobel Edge detection

Sobel edge detection is a popular technique in image processing and computer vision used to detect edges in digital images. It is based on convolving the image with Sobel kernels to compute gradients, which are then used to highlight areas of rapid intensity change, typically corresponding to edges in the image.

The Sobel operators consist of a pair of 3x3 kernels, one for detecting changes in the horizontal direction (usually denoted as the Sobel-X operator) and one for detecting changes in the vertical direction (Sobel-Y operator). These kernels are as follows:

Sobel-X Kernel:
```
-1 0 1
-2 0 2
-1 0 1
```

Sobel-Y Kernel:
```
-1 -2 -1
0  0  0
1  2  1
```

Here's how Sobel edge detection works:

1. **Grayscale Conversion**: To perform edge detection, the first step is often to convert the color image to grayscale. This simplifies the problem to working with a single channel representing intensity.

2. **Convolution with Sobel Kernels**: The grayscale image is then convolved with the Sobel-X and Sobel-Y kernels separately to compute the horizontal and vertical gradients, respectively.

3. **Gradient Magnitude**: The gradient magnitude is calculated for each pixel as the square root of the sum of the squares of the Sobel-X and Sobel-Y gradients. This magnitude represents the strength of the edge at each pixel.

4. **Edge Orientation**: The gradient direction or orientation can also be calculated from the Sobel-X and Sobel-Y gradients. This information can be useful for tasks such as edge thinning or line tracing.

5. **Thresholding**: To identify edges, you can apply a threshold to the gradient magnitude. Pixels with gradient magnitudes above a certain threshold are considered part of an edge.

Sobel edge detection is a simple and effective technique for detecting edges in images. It's widely used in applications such as object detection, image segmentation, and feature extraction in computer vision. In practice, you can implement Sobel edge detection using image processing libraries like OpenCV, which provides functions for applying Sobel operators to images.
